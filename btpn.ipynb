## Dataset Overview
Dataset ini merupakan data asli Bank BTPN Syariah yang di dapat dari program magang VIX Rekamin sebagai data Engineer. Project akhirnya ditugaskan untuk menganalisis penyebab pelanggan yang meninggalkan layanan dan diberikan dtaset sebagai berikut:

### 1. Customer_data_history
- CLIENTNUM = nomor clinet
- idstatus = status client
- Educationid = pendidikan client
- Maritalid = id pernikahan,
- card_categoryid = id jenis kartu
- Gender = Jenis Kelamin
- Dependent_count = tanggungan
- Income_Category = pendapatan
- Months_on_book = jml bulan dari awal buka rekening
- Total_Relationship_Count =  jumlah produk atau layanan keuangan yang dimiliki oleh nasabah di bank
- Months_Inactive_12_mon = jumlah bulan dalam 12 bulan terakhir di mana nasabah tidak aktif menggunakan produk atau layanan keuangan
- Contacts_Count_12_mon = jumlah kontak yang dilakukan oleh nasabah dengan bank dalam jangka waktu 12 bulan terakhir.
- Credit_Limit = jumlah maksimum kredit atau pinjaman yang diberikan oleh bank kepada nasabah
- Total_Revolving_Bal = umlah saldo yang masih harus dibayar oleh nasabah pada akhir periode tagihan kartu kredit atau garis kredit yang terus berputar.
- Avg_Open_To_Buy = jumlah rata-rata kredit yang tersedia bagi nasabah untuk digunakan pada kartu kredit atau garis kredit mereka
- Total_Trans_Amt =  jumlah total transaksi keuangan yang dilakukan oleh nasabah dalam jangka waktu tertentu
- Total_Trans_Ct = jumlah total transaksi yang dilakukan oleh nasabah dalam jangka waktu tertentu.
- Avg_Utilization_Ratio = persentase penggunaan kredit atau garis kredit nasabah dibandingkan dengan batas kredit yang diberikan oleh bank.

### 2. category_db
- id = nomor id kategori
- card_category = jenis kategori kartu

### 3. education_db
- id = nomor id pendidikan
- education_level = tingkat pendidikan

### 4. marital_db
- id = nomor id pernikahan
- Marital_Status = status pernikahan

### 5. status_db
- id = nomor id status kartu
- status = status aktif kartu

## Import library

```python
import pandas as pd
import numpy as np
```

## Load Dataset

```python
# tampilkan semua kolom
pd.set_option('display.max_columns', None)

#read csv
data1 = pd.read_csv('customer_data_history.csv')
data2 = pd.read_csv('category_db.csv')
data3 = pd.read_csv('education_db.csv')
data4 = pd.read_csv('marital_db.csv')
data5 = pd.read_csv('status_db.csv')
```

```python
#show data inti
data1.head()
```

```python
#join data
df_btpn = pd.merge(data1,data2, left_on='card_categoryid',right_on='id')
df_btpn = pd.merge(df_btpn,data3, left_on='Educationid',right_on='id')
df_btpn = pd.merge(df_btpn,data4, left_on='Maritalid',right_on='id')
df_btpn = pd.merge(df_btpn,data5, left_on='idstatus',right_on='id')
```

```python
#show 5 data teratas
df_btpn.head()
```

```python
#show 5 data terbawah
df_btpn.tail()
```

## Preprocessing Data

```python
#cek jumlah baris dan kolom
baris,kolom = df_btpn.shape
print(baris, 'baris')
print(kolom, 'kolom')
```

```python
#cek nama kolom
df_btpn.columns
```

```python
#hapus kolom yang tidak diperlukan
df_btpn.drop(['CLIENTNUM','idstatus','Educationid', 'Maritalid',
           'card_categoryid','id_x','id_y', 'Total_Trans_Amt', 'Total_Trans_Ct',
           'Avg_Utilization_Ratio'],axis=1,inplace=True)
```

```python
#cek info dataset
df_btpn.info()
```

```python
#cek jumlah missing value
df_btpn.isnull().sum()
```

Dan setelah melakukan cek missing value, saya kira datset yang lumayan besar agak aneh tidak memiliki missing value. Selanjutnya kita akan analisis lebih lanjut

```python
#kelompokan data sesuai kategori
data_num = list(df_btpn.select_dtypes(include=['int64','float64']))
data_cat = list(df_btpn.select_dtypes(include=['object']))
```

```python
#cek data numerik
data_num
```

```python
#cek data kategorik
data_cat
```

```python
#cek data unik
for col in df_btpn:
    print('Nama Kolom: ', col.upper())
    print('Nilai Unik: ', df_btpn[col].unique())
    print(100*'=')
```

Setelah melakukan pengecekan nilai unik terdapat niali **Unknown** yang ditermasuk dalam kategori missing value

```python
#cek missing value nilai numerik
df_btpn[
    data_num
].head()
```

Untuk data numerikal tidak ada data aneh,sehingga dapat dipastikan data numerikal sudah bersih

```python
#ubah nilai unknown menjadi NaN
df_btpn.replace('Unknown', np.nan, inplace=True)
```

```python
#cek kembali missing value
df_btpn.isnull().sum()
```

```python
#handle missing value kategorik dengan modus
for col in data_cat:
    df_btpn[col].fillna(df_btpn[col].mode()[0], inplace=True)

#cek kembali missing value
df_btpn.isnull().sum()
```

```python
df_btpn.head()
```

```python
#cek data duplikat
df_btpn.duplicated().sum()
```

```python
df_btpn.rename(columns={'Customer_Age':'Umur',
                        'Gender': 'Jenis_Kelamin',
                        'Dependent_count': 'Jumlah_Tanggungan',
                        'Income_Category': 'Kategori_Pendapatan',
                        'Months_on_book': 'Lama_Berlangganan',
                        'Total_Relationship_Count': 'Jumlah_Relasi',
                        'Months_Inactive_12_mon': 'Bulan_Inaktif_12_bulan',
                        'Contacts_Count_12_mon': 'Jumlah_Dihubungi_12_bulan',
                        'Credit_Limit': 'Limit_Kredit',
                        'Total_Revolving_Bal': 'Total_Saldo_Revolving',
                        'Avg_Open_To_Buy': 'Rata2_Kredit',
                        'Card_Category': 'Kategori_Kartu',
                        'Education_Level': 'Pendidikan',
                        'Marital_Status': 'Pernikahan'}, inplace=True)
```
